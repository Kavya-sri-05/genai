```python
# Step 1: Install HuggingFace transformers library
!pip install transformers gradio

from transformers import pipeline
import gradio as gr

# Step 2: Initialize HuggingFace GPT-2 model
chatbot = pipeline("text-generation", model="gpt2")

# Step 3: Define a function to handle user input and generate responses
def chat_with_bot(question):
    response = chatbot(question, max_length=100, num_return_sequences=1)
    return response[0]['generated_text']

# Step 4: Create the Gradio interface for the chatbot
with gr.Blocks() as chatbot_ui:
    gr.Markdown("## 🤖 AI Chatbot (Powered by HuggingFace GPT-2)")

    # Create the input and output components
    user_input = gr.Textbox(label="Ask me anything!", placeholder="Enter your question here...")
    chatbot_output = gr.Textbox(label="Chatbot's Response")

    # Submit button that calls the chat function
    user_input.submit(chat_with_bot, inputs=user_input, outputs=chatbot_output)

# Step 5: Launch the chatbot UI (with a shareable link)
chatbot_ui.launch(share=True)
```
