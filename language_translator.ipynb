```python
# Install googletrans if not already installed
!pip install googletrans==4.0.0-rc1

# Import the necessary library
from googletrans import Translator, LANGUAGES

# Initialize the translator
translator = Translator()

# Function to translate text
def translate_text(text, src_lang, dest_lang):
    # Translate text from source language to destination language
    translation = translator.translate(text, src=src_lang, dest=dest_lang)
    return translation.text

# Show all available languages
print("Available Languages:")
for code, language in LANGUAGES.items():
    print(f"{code}: {language}")

# Taking user input for text and languages
text_to_translate = input("\nEnter the text you want to translate: ")

# Get source language and destination language from the user
src_language = input("\nEnter the source language code (e.g., 'en' for English, 'es' for Spanish): ")
dest_language = input("Enter the destination language code (e.g., 'en' for English, 'es' for Spanish): ")

# Translate text
translated_text = translate_text(text_to_translate, src_language, dest_language)

# Output the translation
print(f"\nOriginal text: {text_to_translate}")
print(f"Translated text: {translated_text}")
```
